```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Copy of sonar data.csv',header = None)
```

```python
df.sample(5)
```

```python
df[60].value_counts()
```

```python
X = df.drop(columns=df.columns[60],axis=1)
Y = df[df.columns[60]]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
predictions_train = model.predict(X_train)
predictions_test = model.predict(X_test)

```

```python
training_data_accuracy = accuracy_score(predictions_train,Y_train)
```

```python
print(training_data_accuracy)
```

```python
test_data_accuracy = accuracy_score(predictions_test,Y_test)
```

```python
print(test_data_accuracy)
```
